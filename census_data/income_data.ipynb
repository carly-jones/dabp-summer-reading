```python
import pandas as pd
```

```python
var_data = pd.read_csv('census_vars_tract.csv')
distances = pd.read_csv('tract_manhattan_distances.csv')
tracts = pd.read_csv('Allegheny_County_Census_Tracts_2016.csv')['TRACTCE']
```

```python
var_data.columns
```

```python
total_pop = var_data['est_total_pop_u18_in_households_6to8yrs'].sum()
total_pop
```

```python
# var_data['below50k'] = var_data['est_total_household_income_u10k'].astype(int) + var_data['est_total_household_income_10to15k'].astype(int) + var_data['est_total_household_income_15to20k'].astype(int) + var_data['est_total_household_income_20to25k'].astype(int) + var_data['est_total_household_income_25to30k'].astype(int) + var_data['est_total_household_income_30to35k'].astype(int) + var_data['est_total_household_income_35to40k'].astype(int) + var_data['est_total_household_income_40to45k'].astype(int) + var_data['est_total_household_income_45to50k'].astype(int)
# var_data['below50asperc'] = var_data['below50k'] / var_data['est_total_household_income'].astype(int)

var_data['below60k'] = var_data['est_total_household_income_u10k'].astype(int) + var_data['est_total_household_income_10to15k'].astype(int) + var_data['est_total_household_income_15to20k'].astype(int) + var_data['est_total_household_income_20to25k'].astype(int) + var_data['est_total_household_income_25to30k'].astype(int) + var_data['est_total_household_income_30to35k'].astype(int) + var_data['est_total_household_income_35to40k'].astype(int) + var_data['est_total_household_income_40to45k'].astype(int) + var_data['est_total_household_income_45to50k'].astype(int) + var_data['est_total_household_income_50to60k'].astype(int)
var_data['below60asperc'] = var_data['below60k'] / var_data['est_total_household_income'].astype(int)
```

```python
morethan50below60k = var_data[var_data['below60asperc'] > 0.5]
pop = morethan50below60k['est_total_pop_u18_in_households_6to8yrs'].sum()
pop / total_pop
```
